## Import Semua Packages/Library yang Digunakan

```python
## Import Semua Packages/Library yang Digunakan

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
import matplotlib.dates as mdates
```

## Data Wrangling

### Gathering Data

```python
### Gathering Data

# Dataset URL
url = "https://raw.githubusercontent.com/marceloreis/HTI/master/PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv"
data = pd.read_csv(url)



```

**Insight:**
- Dataset diperoleh dari sumber GitHub dan diimpor dalam format CSV untuk analisis lebih lanjut.

### Assessing Data

```python
## Assessing Data
st.header("Data Awal")
st.write("Eksplorasi dataset kualitas udara Beijing dengan fitur interaktif di bawah ini.")
st.write("Contoh 5 baris data:")
st.write(data.head())
st.write("Informasi dataset:")
st.write(data.info())
st.write("Jumlah nilai kosong:")
st.write(data.isnull().sum())
```

**Insight:**
- Evaluasi awal mengidentifikasi keberadaan data tidak lengkap (missing values) yang memerlukan proses pembersihan komprehensif untuk menjamin keakuratan hasil analisis.
- Kumpulan data ini mengandung beragam parameter kunci, meliputi informasi temporal, tingkat konsentrasi polutan PM2.5, serta berbagai parameter meteorologis seperti temperatur (TEMP), tekanan atmosfer (PRES), dan variabel penting lainnya.

### Cleaning Data

```python
# Menghapus nilai kosong
data = data.dropna()

# Gabungkan kolom tahun, bulan, dan hari menjadi satu kolom tanggal
data['date'] = pd.to_datetime(data[['year', 'month', 'day']])
filtered_data = data.copy()
filtered_data['month'] = filtered_data['date'].dt.to_period("M")



```

**Insight:**
- Teridentifikasi pola fluktuasi signifikan pada nilai rata-rata PM2.5 berdasarkan bulan, mengindikasikan variabilitas tingkat polusi udara sesuai dengan perubahan waktu.
- Analisis korelasi antara konsentrasi PM2.5 dengan berbagai faktor meteorologis memberikan pemahaman mendalam tentang determinan utama yang mempengaruhi kualitas udara.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
### Explore Tren Rata-rata PM2.5 Per Bulan

# Membuat kolom periode bulanan
filtered_data = data.copy()
filtered_data['month'] = filtered_data['date'].dt.to_period("M")
monthly_avg_pm25 = filtered_data.groupby('month')['PM2.5'].mean().reset_index()

correlation_matrix = filtered_data[['PM2.5', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']].corr()

# Scatterplot PM2.5 vs TEMP
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='TEMP', y='PM2.5', color='teal', alpha=0.5)
plt.title("Scatterplot: PM2.5 vs Suhu (TEMP)")
plt.xlabel("Suhu (TEMP)")
plt.ylabel("PM2.5")
plt.grid()
plt.show()

# Scatterplot PM2.5 vs Wind Speed (WSPM)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='WSPM', y='PM2.5', color='orange', alpha=0.5)
plt.title("Scatterplot: PM2.5 vs Kecepatan Angin (WSPM)")
plt.xlabel("Kecepatan Angin (WSPM)")
plt.ylabel("PM2.5")
plt.grid()
plt.show()

# Scatterplot PM2.5 vs TEMP
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='TEMP', y='PM2.5', color='teal', alpha=0.5)
plt.title("Scatterplot: PM2.5 vs Suhu (TEMP)")
plt.xlabel("Suhu (TEMP)")
plt.ylabel("PM2.5")
plt.grid()
plt.show()

# Scatterplot PM2.5 vs Wind Speed (WSPM)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='WSPM', y='PM2.5', color='orange', alpha=0.5)
plt.title("Scatterplot: PM2.5 vs Kecepatan Angin (WSPM)")
plt.xlabel("Kecepatan Angin (WSPM)")
plt.ylabel("PM2.5")
plt.grid()
plt.show()

# Heatmap Korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="viridis", fmt=".2f", cbar=True)
plt.title("Heatmap Korelasi Variabel Numerikal")
plt.show()

# Tambahkan kolom kategori bulan
filtered_data['month_name'] = filtered_data['date'].dt.strftime('%B')

# Boxplot PM2.5 berdasarkan Bulan
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_data, x='month_name', y='PM2.5', palette="Set3")
plt.xticks(rotation=45)
plt.title("Distribusi PM2.5 Berdasarkan Bulan")
plt.xlabel("Bulan")
plt.ylabel("PM2.5")
plt.grid()
plt.show()

# Boxplot PM2.5 berdasarkan Jam
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_data, x='hour', y='PM2.5', palette="Set2")
plt.title("Distribusi PM2.5 Berdasarkan Jam")
plt.xlabel("Jam")
plt.ylabel("PM2.5")
plt.grid()
plt.show()

# Pairplot untuk variabel numerikal
numerical_features = ['PM2.5', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
sns.pairplot(filtered_data[numerical_features], diag_kind='kde', corner=True)
plt.show()
```

**Insight:**
- Visualisasi tren temporal menampilkan dinamika peningkatan dan penurunan konsentrasi PM2.5 yang substansial selama interval waktu yang dianalisis.
- Pemetaan korelasi melalui heatmap secara komprehensif mengilustrasikan interaksi kompleks dan hubungan interdependensi antar variabel dalam dataset.

## Visualization & Explanatory Analysis



### Pertanyaan 1: Tren Rata-rata PM2.5 Per Bulan

```python
## Pertanyaan 1: Tren Rata-rata PM2.5 Per Bulan
st.header("Tren Rata-rata PM2.5 Per Bulan")
view_option = st.radio("Pilih Tampilan:", ["Grafik", "Tabel Data"], horizontal=True)
if view_option == "Tabel Data":
    st.write(monthly_avg_pm25)
else:
    monthly_avg_pm25 = filtered_data.groupby('month')['PM2.5'].mean().reset_index()
    monthly_avg_pm25['month'] = monthly_avg_pm25['month'].dt.to_timestamp()
    monthly_avg_pm25['month'] = monthly_avg_pm25['month'].dt.strftime('%Y-%m')

# Visualisasi Tren Bulanan
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_pm25['month'], monthly_avg_pm25['PM2.5'], marker='o', color='purple')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Set interval label
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Format tahun-bulan
plt.xticks(rotation=45)
plt.title("Rata-rata PM2.5 per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Rata-rata PM2.5")
plt.tight_layout()
st.pyplot(plt.gcf())

# Menghitung rata-rata PM2.5 per bulan
monthly_avg_pm25 = filtered_data.groupby('month')['PM2.5'].mean().reset_index()
monthly_avg_pm25['month'] = monthly_avg_pm25['month'].dt.to_timestamp()

# Visualisasi menggunakan matplotlib
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_pm25['month'], monthly_avg_pm25['PM2.5'], marker='o', color='purple')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.title("Tren Rata-rata PM2.5 per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Rata-rata PM2.5")
plt.grid()
plt.show()
```

### Pertanyaan 2:

```python
## Pertanyaan 2: Korelasi PM2.5 dengan Faktor Cuaca
st.header("Korelasi PM2.5 dengan Variabel Cuaca")
correlation_threshold = st.slider("Pilih Ambang Korelasi:", min_value=0.0, max_value=1.0, value=0.0, step=0.1)
correlation_filtered = correlation_matrix.where(abs(correlation_matrix) >= correlation_threshold)
correlation_matrix = filtered_data[['PM2.5', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="viridis", cbar=True)
plt.title("Korelasi PM2.5 dengan Variabel Cuaca")
st.pyplot(plt.gcf())

# Memilih kolom yang akan dianalisis
columns_of_interest = ['PM2.5', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
correlation_matrix = filtered_data[columns_of_interest].corr()

# Visualisasi korelasi menggunakan heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="viridis", cbar=True)
plt.title("Korelasi PM2.5 dengan Variabel Cuaca")
plt.show()
```

**Insight:**
- Analisis distribusi PM2.5 mengungkapkan karakteristik sebaran data yang memberikan konteks penting untuk interpretasi hasil.
- Pola variasi konsentrasi PM2.5 berdasarkan jam menunjukkan ritme diurnal yang konsisten, mencerminkan pengaruh aktivitas manusia dan faktor lingkungan terhadap kualitas udara.

## Analisis Lanjutan (Opsional)

```python
# Distribusi PM2.5
st.header("Distribusi PM2.5")
bin_count = st.slider("Jumlah Bins:", min_value=10, max_value=100, value=30, step=5)
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data['PM2.5'], kde=True, bins=30, color='darkviolet')
plt.title("Distribusi PM2.5")
plt.xlabel("PM2.5")
plt.ylabel("Frekuensi")
st.pyplot(plt.gcf())

# Rata-rata PM2.5 Berdasarkan Jam
st.header("Rata-rata PM2.5 Berdasarkan Jam")
chart_type = st.selectbox("Pilih Jenis Grafik:", ["Line", "Bar"])
hourly_avg_pm25 = filtered_data.groupby('hour')['PM2.5'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(hourly_avg_pm25['hour'], hourly_avg_pm25['PM2.5'], marker='o', color='purple')
plt.title("Rata-rata PM2.5 Berdasarkan Jam")
plt.xlabel("Jam")
plt.ylabel("Rata-rata PM2.5")
plt.tight_layout()
st.pyplot(plt.gcf())
```

## Conclusion

 - Pertanyaan 1 tentang tren rata-rata PM2.5:

Tingkat PM2.5 bervariasi sepanjang tahun, dengan nilai tertinggi pada bulan-bulan musim dingin, yang berlangsung dari Desember hingga Februari.

 - Pertanyaan 2 yang berkaitan dengan korelasi dengan faktor cuaca:

PM2.5 memiliki korelasi positif dengan kelembapan (DEWP) dan korelasi negatif dengan kecepatan angin (WSPM), yang menunjukkan bahwa angin membantu menyebarkan polutan. Kedua korelasi menunjukkan bahwa kelembapan tinggi dapat memperburuk kondisi polusi udara.
